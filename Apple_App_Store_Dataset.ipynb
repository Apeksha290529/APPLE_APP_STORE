```python
df['type']=np.where(df['Free']==True,"Free","Paid")
```

```python
df['Content_Rating'] = df['Content_Rating'].replace("4+","Preschoolers")
df['Content_Rating'] = df['Content_Rating'].replace("9+","Older-Children")
df['Content_Rating'] = df['Content_Rating'].replace("12+","Teen")
df['Content_Rating'] = df['Content_Rating'].replace("17+","Adults")
df['Content_Rating'] = df['Content_Rating'].replace("Not yet rated","Every-One")
```

```python
df['Content_Rating'].unique()
```

```python
def categorize_reviews(review_count):
    if 0 <= review_count < 10000:
        return 'Less than 10K'
    elif 10000 <= review_count < 500000:
        return 'Between 10K and 500K'
    elif 500000 <= review_count < 1000000:
        return 'Between 500K and 1Mil'
    elif 1000000 <= review_count <= 22685334:
        return 'Million Plus'

df['review_category'] = df['Reviews'].apply(categorize_reviews)
```

```python
df['Released'] = pd.to_datetime(df['Released'])
df['Year'] = df['Released'].dt.year
app_counts = df['Year'].value_counts().sort_index()
app_counts_df = pd.DataFrame({'Year':app_counts.index, 'Number of Apps':app_counts.values})
app_counts_df.style.background_gradient(subset='Number of Apps', cmap='Blues')
```

```python
app_counts = df['Developer'].value_counts().sort_values(ascending=False)
top_10_developers = app_counts[:10]
plt.figure(figsize=(10,5))
plt.barh(top_10_developers.index, top_10_developers.values, color='skyblue')
plt.xlabel('Number of Apps')
plt.title('Top 10 Developers Producing The Most Apps')
plt.gca().invert_yaxis()
plt.show()
```

```python
top_10_developers_df = pd.DataFrame({'Developer':top_10_developers.index, 'Number of Apps':top_10_developers.values})
top_10_developers_df
```

```python
high_rating_apps = df[df['Reviews'] > 100000]
category_counts = high_rating_apps['Primary_Genre'].value_counts()
plt.figure(figsize=(10,5))
plt.barh(category_counts.index, category_counts.values, color='skyblue')
plt.xlabel('Number of Apps')
plt.title('Distribution of Apps with Rating Count Greater Than 100,000 Across Categories')
plt.gca().invert_yaxis()
plt.show()
```
